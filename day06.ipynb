```python
#i 0으로 초기화
i = 0
#sum 0으로 초기화
sum = 0
#i가 100보다 작을때까지 반복
while i < 100 :
    #1씩 더해주고
    i += 1
    #sum에다가 누적한다
    sum = sum + i
print("sum = ",sum)
```

```python
#i는 1로 초기화
i=1
#무한 반복
while True:
    #i를 j에 담는다
    j=i
    while True:
        #만약 j가 10이라면 탈출
        if(j == 10) : 
            break
        print( "*",end=  "")
        #j를 1씩 증감
        j+=1
    #i 가 10이 되면 탈출
    if(i == 10) : 
        break
    print(end = "\n")
    #i 1씩 증감
    i+=1
```

```python
#Mon 에 1저장
Mon=1
#각각 월의 마지막 날짜를 저장
for LastDay in [31,28,31,30,31,30,31,31,30,31,30,31]:
    print(Mon,"월 :",end="")
    #배열의 순서대로 가져와서 그 값만큼 반복돌려 출력
    for Day in range(1,LastDay+1):
        print(Day,end=",")
    print(end="\n")
    Mon+=1
```

```python
#2부터 9까지 3씩 증감 반복
for i in range(2,10,3):
    #2부터 9까지 반복
    for j in range(2,10):
        print(i,"*",j,"=",i*j,end="\t")
        print(i+1,"*",j,"=",i*j,end="\t")
        #아래까지 출력하고 줄바꿈
        if(i+2!=10): 
            print(i+2,"*",j,"=",i*j,end="\n")
        else: print(end="\n")
    print(end="\n")
```

```python
for i in range(2,10,3):
    for j in range(2,10):
        print(i,"*",j,"=",i*j,end="\t")
        print(i+1,"*",j,"=",(i+1)*j,end="\t")
        if(i+2!=10): 
            print(i+2,"*",j,"=",(i+2)*j,end="\n")
        else: print(end="\n")
    print(end="\n")
```

```python
#i 가2부터 4까지 반복 총 3번 반복
for i in range(2,5):
    #j를 2부터 9까지 반복
    for j in range(2,10):
        #처음은 i곱하고 i+3을 하고 i+6을 곱해서 출력한다.
        print(i,"*",j,"=",i*j,end="\t")
        print(i+3,"*",j,"=",(i+3)*j,end="\t")
        #조건문에 10이되면 끝난다.
        if(i+6 != 10): print(i+6,"*",j,"=",(i+6)*j,end="\n")
        else: print(end="\n")
    print(end="\n")
```
