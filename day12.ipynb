```python
f=0
#발생되는 에러에 대하여 예외 에러 처리(try / except / finally)를 하여야 함
try:
    #open을 함수를 통해 LICENSE.txt를 읽는다.
    f=open("LICENSE.txt","rt")
    #반복
    while True:
        #한줄씩 읽어서 r에 담는다
        r=f.readline()
        #r이 값이 없으면 break
        if not r:
            break          
        print(r)
        
except FileNotFoundError:
    print("파일없음")
finally:
    if(f): f.close()
    print("프로그램 끝")
```

```python
#random 임포트 r로 참조
import random as r
#open()함수를 통해 score.txt에다 쓴다
#파일이 쓸때는 저절로 만들어진다. 그걸 f에 담는다.
f=open("score.txt","wt")
#100번 반복
for i in range(100):
    #이름과 0~100랜덤 점수를 각각 담아 one_person에 담는다.
    one_person="홍길동"+str(i)+","+str(r.randrange(0,100)) +","+str(r.randrange(0,100))+","+str(r.randrange(0,100))+"\n"
    f.write(one_person) #한 줄씩 쓴다
    print(one_person)
    
#닫아준다.   
f.close()
```

```python
#open()함수를 통해 score.txt 읽어서 f에 담는다.
f=open("score.txt","rt")
#반복
while True:
    one_person=f.readline() #한 줄씩 읽는다
    if not one_person: 
        break #읽을 내용이 없으면 루프탈출

    one_person=one_person.replace("\n","") #줄 바꿈문자 없애기
    item=one_person.split(",") #콤마 구분자로 짜르기
    print("="*20)
    print("이름: ", item[0])
    print("국어: ", item[1])
    print("영어: ", item[2])
    print("수학: ", item[3])
    print("="*20)
    
f.close()
```

```python
conn = sqlite3.connect("mydata.db")
cur = conn.cursor()
cur.execute("SELECT * FROM Twice_Table")

rows = cur.fetchall()

for field in rows: #해당 데이터를 반복문을 통해 하나씩 처리
    kor=field[2]
    eng=field[3]
    mat=field[4]
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d"%(field[0],field[1],kor,eng,mat,kor+eng+mat,int((kor+eng+mat)/3)))
    
cur.close()
conn.close()
```
