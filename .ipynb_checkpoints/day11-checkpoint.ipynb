```python
from math import sin,pi

for i in range(360):
    sin_value=sin(i*pi/180)
    space_value=int( (1+sin_value)*50)
    print(" "*space_value, end="")
    print("[%f]"%sin_value)
```

```python
from math import sin,pi
import matplotlib.pyplot as plt

graph_value=[] #리스트 선언,빈 리스트를 만들어야 append가능

for i in range(360):
    graph_value.append(sin(i*pi/180)) #sin값을 하나씩 리스트에 넣음
    
plt.plot(graph_value) #해당 리스트를 그래프 함수에 보냄
plt.show()
```

```python
#통계 관련 모듈
import statistics as s

a=[100,95,25,33,45,90,100,35,25]
print(s.mean(a)) #평균
print(s.stdev(a)) #표준편차
print(s.variance(a)) #분산
```

```python
#날짜와 시간 모듈
import time as t
#현재시간을 구함
a=t.time();
print(a) 
#초단위로 표시된 시간을 문자형식으로 표현
t.ctime(a) 
#우리나라 기준으로 time=0의 경우 1970년1월1일 오전9시 정각이다
print(t.ctime(0))
print(t.ctime())
```

```python
#time 모듈 임포트
import time as t
#localtime()함수 시간 요소를 구조체로 반환함
#time에 localtime()현재 시간을 가져와 a저장
a = t.localtime()
print(a)
print()
#a안에는 년월일시간이 있어서 a로접근해서 출력
print("%d년 %d월 %d일 %02d시 %02d분 %02d초"%(a.tm_year,a.tm_mon,a.tm_mday,a.tm_hour,a.tm_min,a.tm_sec))
```

```python
import datetime as t
# datetime 모듈의 now()함수로 보다 직관적으로 시간을 출력할 수 있음
a=t.datetime.now()
print("%d년 %d월 %d일 %02d시 %02d분 %02d초"%(a.year,a.month,a.day,a.hour,a.minute,a.second))
```

```python
import time as t
#현재 시간을 가져와 start에 담는다.
start=t.time()
for i in range(10000):
    pass
#현재 시간을 가져와 end에 담는다.
end=t.time()
#end - start를 하면 반복실행시간을 구할수 있다.
print("%f 초 동안 실행함"%(end-start))
```

```python
import time as t
#10번 반복
for i in range(10):
    #현재 시간 출력
    print(t.ctime())
    #1초 재운다
    t.sleep(1)
```

```python
#random모듈 matplotlib모듈 임포트
import random as r
import matplotlib.pyplot as plt
#리스트 생성
score=[]
#100번 반복
for i in range(100):
    #score리스트에 값을 추가 랜덤숫자 0~100
    score.append(r.randrange(0,100))
#그래프로 만듬  
plt.plot(score)
plt.show()
```

```python
# operating system(운영체계 윈도우)을 제어 가능
import os
#os.getcwd()로 현재 실행 디렉토리를 얻을 수 있음
os.getcwd()
#os.system(명령)으로 윈도우의 명령을 실행할 수 있음
os.system("calc") #계산기 실행
os.system("python") #파이썬 실행 실행된 상태에서 기다림
os.system("shutdown -f -t 10") #이 명령 수행전 모든 윈도우 프로그램을 정리할 것
```

```python
#urllib을 이용하여 홈페이지 소스를 분석
from urllib.request import urlopen
#urlopen 함수를 이용하여 학교 홈페이지('http://www.kopo.ac.kr/')
#중 첫 페이지의 소스를 한 줄씩 받아옴
with urlopen('http://www.kopo.ac.kr/') as response:
    for line in response:
        line = line.decode('utf-8') # 한글 처리
        #만약 해당 줄에 폴리텍이 있으면 해당줄 출력
        if "폴리텍" in line:
            print(line)
```
